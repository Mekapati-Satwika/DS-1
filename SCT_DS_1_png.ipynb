```python
# Install required libraries
import matplotlib.pyplot as plt
import numpy as np

# Data: India's Population by Age Group (2022)
age_groups = ['0-20 Years', '21-64 Years', '65+ Years']
population = [512, 807, 98]  # in millions
colors = ['#FFD700', '#4169E1', '#FF69B4']

# Create Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, population, color=colors, edgecolor='black', width=0.5)

# Add value labels on each bar
for bar, val in zip(bars, population):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 10,
             f'{val} Mn', ha='center',
             fontsize=12, fontweight='bold')

# Labels and Title
plt.title("India's Population Distribution by Age Group (2022)",
          fontsize=14, fontweight='bold')
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population (in Millions)", fontsize=12)
plt.ylim(0, 950)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('SCT_DS_1.png', dpi=150)
plt.show()

print("Chart saved successfully!")
```
